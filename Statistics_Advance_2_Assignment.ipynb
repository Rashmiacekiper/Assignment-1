## Q 1.  What is hypothesis testing in statistics?

Ans- Hypothesis testing in statistics is a method used to make inferences or draw conclusions about a population based on sample data. It involves testing an assumption (the hypothesis) about a population parameter, such as the mean or proportion. The goal is to determine whether there is enough evidence in the sample data to support or reject the hypothesis.

Here’s a simple breakdown of the process:

1. **Formulate the hypotheses**:  
   - **Null hypothesis (H₀)**: This is a statement of no effect, no difference, or no change. It represents the status quo or the assumption you're testing.
   - **Alternative hypothesis (H₁ or Ha)**: This is the opposite of the null hypothesis. It suggests that there is a difference or effect.

2. **Set the significance level (α)**:  
   This is the probability of rejecting the null hypothesis when it is actually true, often set at 0.05 (5%).

3. **Collect the data**:  
   Gather sample data from the population to test the hypothesis.

4. **Conduct the test**:  
   Use statistical methods (like t-tests, z-tests, chi-squared tests, etc.) to calculate a test statistic, which measures the degree to which the sample data supports the null hypothesis.

5. **Determine the p-value**:  
   The p-value tells you the probability of observing the data (or something more extreme) if the null hypothesis were true. A small p-value (typically ≤ 0.05) indicates strong evidence against the null hypothesis.

6. **Make a decision**:  
   - If the p-value is less than or equal to the significance level (α), reject the null hypothesis. This means there is enough evidence to support the alternative hypothesis.
   - If the p-value is greater than α, fail to reject the null hypothesis. This means there isn't enough evidence to support the alternative hypothesis.

### Example:
Imagine you're testing whether a new drug is effective in lowering blood pressure. Your null hypothesis (H₀) would be that the drug has no effect (i.e., the mean change in blood pressure is 0), while the alternative hypothesis (H₁) would suggest the drug does have an effect.

The outcome of hypothesis testing helps to make informed decisions, but it does not provide absolute proof; rather, it shows the strength of the evidence available.

## Q 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

Ans- The **Z-score** is calculated using the formula:

\[
Z = \frac{(X - \mu)}{\sigma}
\]

Where:
- **X** = sample value (or sample mean)
- **μ** = population mean
- **σ** = population standard deviation

### What does it represent?
The Z-score measures how many standard deviations a data point (or sample mean) is away from the population mean.

- A **positive Z-score** means the value is above the mean.
- A **negative Z-score** means the value is below the mean.

In hypothesis testing, the Z-score helps determine how extreme the observed data is relative to the null hypothesis. A larger absolute value of the Z-score indicates stronger evidence against the null hypothesis.

## Q 16. How is Bayes' Theorem used in statistics, and what is its significance?
Ans- **Bayes' Theorem** is a fundamental concept in statistics that provides a way to update the probability of a hypothesis based on new evidence or data. It combines prior knowledge (the initial probability) with new information to refine predictions.

### Formula:
\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]
Where:
- **P(A|B)** is the **posterior probability** (the probability of A given B).
- **P(B|A)** is the **likelihood** (the probability of B given A).
- **P(A)** is the **prior probability** (the initial probability of A).
- **P(B)** is the **marginal likelihood** (the total probability of B).

### Significance:
Bayes' Theorem allows you to:
- Update beliefs or probabilities as new evidence becomes available.
- Make more informed decisions, especially in situations with uncertainty.

In short, it provides a powerful way to adjust probabilities and make predictions based on both prior knowledge and new data.

## Q 18. What is the Chi-square goodness of fit test, and how is it applied?
Ans- The **Chi-square goodness of fit test** is used to determine whether the observed frequency distribution of a categorical variable matches the expected distribution based on a specific hypothesis.

### How is it applied?
1. **Formulate hypotheses**:
   - **Null hypothesis (H₀)**: The observed data fits the expected distribution.
   - **Alternative hypothesis (H₁)**: The observed data does not fit the expected distribution.
   
2. **Calculate the test statistic**:  
   \[
   \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
   \]
   Where:
   - \(O_i\) = observed frequency for category \(i\)
   - \(E_i\) = expected frequency for category \(i\)

3. **Determine the degrees of freedom (df)**:  
   \(df = \text{number of categories} - 1\)

4. **Compare with the critical value** from the Chi-square distribution table. If the calculated chi-square statistic is greater than the critical value at a given significance level, reject the null hypothesis.

In short, the Chi-square goodness of fit test is used to assess if observed data matches the expected distribution, helping to evaluate categorical data patterns.

## Q 20. What is an ANOVA test, and what are its assumptions?
Ans- The **ANOVA (Analysis of Variance)** test is a statistical method used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others.

### Assumptions of ANOVA:
1. **Independence**: The samples from each group are independent of each other.
2. **Normality**: The data in each group are approximately normally distributed.
3. **Homogeneity of variances**: The variances of the groups being compared are approximately equal (also called **homoscedasticity**).

In short, ANOVA is used to test for significant differences between group means, and its validity depends on the independence of samples, normality, and equal variances across groups.


## Q 21. What are the different types of ANOVA tests?
Ans- There are several types of **ANOVA tests**, each used for different scenarios:

1. **One-way ANOVA**:
   - Used to compare the means of **three or more independent groups** based on a single factor (independent variable).
   - Example: Comparing test scores of students from different teaching methods.

2. **Two-way ANOVA**:
   - Used to examine the effect of **two independent factors** on a dependent variable, and also to explore if there is an interaction between the two factors.
   - Example: Studying the effects of both diet type and exercise on weight loss.

3. **Repeated Measures ANOVA**:
   - Used when the same subjects are measured multiple times (e.g., before and after an intervention), typically in longitudinal studies.
   - Example: Comparing blood pressure of individuals before and after taking a medication over several months.

4. **Multivariate ANOVA (MANOVA)**:
   - Extends ANOVA to handle **multiple dependent variables** simultaneously.
   - Example: Examining the impact of a treatment on both weight and blood pressure.

In short, the different types of ANOVA are used depending on the number of independent variables, the structure of the data (repeated measures), and whether multiple dependent variables are involved.

## Q 1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
import scipy.stats as stats

# Sample data (replace with your own)
sample_data = [20, 22, 23, 24, 25, 21, 20, 23, 24, 22]  # Sample data
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
sample_std_dev = np.std(sample_data, ddof=1)  # Sample standard deviation

# Known population parameters
population_mean = 22  # Known population mean
population_std_dev = 3  # Known population standard deviation

# Z-test formula: Z = (sample_mean - population_mean) / (population_std_dev / sqrt(sample_size))
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate the p-value for a two-tailed test (assuming a significance level of 0.05)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

## Q 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05):
    # Calculate the Z-score for the sample mean
    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Critical Z-values for two-tailed test
    z_critical_left = stats.norm.ppf(alpha / 2)
    z_critical_right = stats.norm.ppf(1 - alpha / 2)

    # Type 1 error (α) - probability of rejecting H₀ when H₀ is true
    type_1_error = alpha

    # Type 2 error (β) - probability of failing to reject H₀ when H₀ is false
    # Assume true population mean is 105 (alternative hypothesis: H₁: μ ≠ 100)
    true_population_mean = 105
    z_score_true = (true_population_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))
    type_2_error = stats.norm.cdf(z_critical_left - z_score_true) + (1 - stats.norm.cdf(z_critical_right - z_score_true))

    # Print Type 1 and Type 2 errors
    print(f"Type 1 Error (α): {type_1_error}")
    print(f"Type 2 Error (β): {type_2_error}")

    # Visualization of the Z-distribution and error regions
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    plt.plot(x, y, label='Standard Normal Distribution')

    # Type 1 error (α) regions (rejection regions)
    plt.fill_between(x, y, where=(x <= z_critical_left) | (x >= z_critical_right), color='red', alpha=0.3, label="Type 1 Error (α) Region")

    # Type 2 error (β) region (fail to reject the null hypothesis)
    plt.fill_between(x, y, where=(x > z_critical_left) & (x < z_critical_right), color='blue', alpha=0.3, label="Type 2 Error (β) Region")

    # Mark the critical values (rejection regions)
    plt.axvline(z_critical_left, color='green', linestyle='--', label=f"Critical Z Left: {z_critical_left:.2f}")
    plt.axvline(z_critical_right, color='green', linestyle='--', label=f"Critical Z Right: {z_critical_right:.2f}")

    # Mark the sample Z-score
    plt.axvline(z_score, color='purple', linestyle='--', label=f"Sample Z-score: {z_score:.2f}")

    # Labels and title
    plt.title("Type 1 and Type 2 Error Visualization")
    plt.xlabel("Z-value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()

# Example usage
sample_mean = 103  # Sample mean
population_mean = 100  # Population mean under H₀
population_std_dev = 10  # Population standard deviation
sample_size = 30  # Sample size

visualize_errors(sample_mean, population_mean, population_std_dev, sample_size)
```

## Q 6. Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy import stats

# Sample data for two independent groups
group_1 = [22, 23, 21, 24, 25, 23, 22, 21, 24, 23]
group_2 = [30, 31, 29, 32, 34, 33, 30, 31, 32, 30]

# Perform the independent T-test
t_stat, p_value = stats.ttest_ind(group_1, group_2)

# Output the T-statistic and p-value
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two groups.")
```

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    # Critical value for the confidence level (t-distribution)
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha / 2, df=sample_size - 1)  # t-critical value for two-tailed

    # Margin of error
    margin_of_error = t_critical * (sample_std_dev / np.sqrt(sample_size))

    # Confidence interval
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

    return confidence_interval

# Example usage
sample_data = [102, 98, 105, 110, 95, 99, 100, 107, 101, 103]
confidence_level = 0.95

confidence_interval = calculate_confidence_interval(sample_data, confidence_level)

print(f"The {confidence_level*100}% confidence interval for the sample mean is: {confidence_interval}")
```

## Q 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process


```python
def bayes_theorem(prior, likelihood, evidence):
    """
    Bayes' Theorem: P(A|B) = (P(B|A) * P(A)) / P(B)

    prior: P(A) - prior probability of hypothesis A
    likelihood: P(B|A) - likelihood of evidence B given hypothesis A
    evidence: P(B) - total probability of the evidence
    """
    # Applying Bayes' Theorem
    posterior = (likelihood * prior) / evidence
    return posterior

# Example values
prior = 0.4        # Prior probability of the hypothesis A (e.g., disease presence)
likelihood = 0.9   # Likelihood of the evidence B given A (e.g., test positive given disease)
evidence = 0.7     # Probability of evidence (e.g., test positive overall)

# Calculate the posterior probability
posterior = bayes_theorem(prior, likelihood, evidence)

print(f"The posterior probability is: {posterior}")
```

## Q 16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    # Calculate the sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # F-statistic: larger variance / smaller variance
    f_stat = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Calculate the p-value for the F-statistic
    p_value = 1 - stats.f.cdf(f_stat, df1, df2)

    return f_stat, p_value

# Example data (two random samples)
sample1 = np.random.normal(0, 1, 100)  # Sample 1 with variance 1
sample2 = np.random.normal(0, 2, 100)  # Sample 2 with variance 4

# Perform the F-test
f_stat, p_value = f_test(sample1, sample2)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

## Q 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    # 1. Normality: Use the Shapiro-Wilk test for normality
    normality_results = [stats.shapiro(group) for group in groups]
    normality_p_values = [result[1] for result in normality_results]

    # 2. Equal Variance: Use Levene's test for equal variances
    levene_stat, levene_p_value = stats.levene(*groups)

    # 3. Independence: This is typically assumed by design, but we can visualize it using scatter plots
    #    Assuming that samples are independent, we don't perform a statistical test for this assumption.

    # Normality assumption check
    normality_pass = all(p > 0.05 for p in normality_p_values)  # If p > 0.05, the group is normally distributed

    # Equal variance check
    equal_variance_pass = levene_p_value > 0.05  # If p > 0.05, we fail to reject the null hypothesis (equal variance)

    # Display results
    print("Normality Assumption Check (Shapiro-Wilk Test):")
    for i, p_value in enumerate(normality_p_values):
        print(f"Group {i+1}: p-value = {p_value} -> {'Normal' if p_value > 0.05 else 'Not Normal'}")

    print(f"\nEqual Variance Check (Levene's Test): p-value = {levene_p_value} -> {'Equal Variance' if equal_variance_pass else 'Not Equal Variance'}")

    # Visualization of normality using Q-Q plot for each group
    plt.figure(figsize=(10, 6))
    for i, group in enumerate(groups):
        plt.subplot(1, len(groups), i+1)
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Group {i+1} Q-Q Plot")

    plt.tight_layout()
    plt.show()

    # Return the results
    return normality_pass, equal_variance_pass

# Example data: Three groups with different means
group1 = np.random.normal(10, 2, 30)  # Group 1 (mean = 10)
group2 = np.random.normal(12, 2, 30)  # Group 2 (mean = 12)
group3 = np.random.normal(14, 2, 30)  # Group 3 (mean = 14)

# Check ANOVA assumptions
normality, equal_variance = check_anova_assumptions(group1, group2, group3)

# Interpretation
if normality and equal_variance:
    print("\nAssumptions met: Normality and Equal Variance for ANOVA.")
else:
    print("\nAssumptions not met for ANOVA.")
```

## Q 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results


```python
import numpy as np
import scipy.stats as stats

# Simulate random data from a normal distribution
np.random.seed(0)
sample_data = np.random.normal(loc=15, scale=3, size=100)  # Mean = 15, Std dev = 3, Sample size = 100

# Known population variance for comparison
population_variance = 9  # Variance = standard deviation^2 = 3^2 = 9

# Sample variance
sample_variance = np.var(sample_data, ddof=1)  # Use ddof=1 for sample variance

# Degrees of freedom (sample size - 1)
df = len(sample_data) - 1

# Chi-square test statistic formula: X² = (n - 1) * s² / σ²
chi_square_statistic = (df * sample_variance) / population_variance

# Calculate p-value (two-tailed test)
# For one-tailed Chi-square test (testing if variance is larger or smaller)
p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))

# Print the results
print(f"Chi-square statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample variance is significantly different from the population variance.")
else:
    print("Fail to reject the null hypothesis: The sample variance is not significantly different from the population variance.")
```

## Q 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats

# Sample data: Number of successes and total observations for two groups
# Group 1: 50 successes out of 200 trials
# Group 2: 30 successes out of 150 trials
successes_group1 = 50
trials_group1 = 200

successes_group2 = 30
trials_group2 = 150

# Proportions for each group
p1 = successes_group1 / trials_group1
p2 = successes_group2 / trials_group2

# Pooled proportion (under the null hypothesis that the proportions are equal)
pooled_p = (successes_group1 + successes_group2) / (trials_group1 + trials_group2)

# Standard error for the difference in proportions
SE = np.sqrt(pooled_p * (1 - pooled_p) * (1 / trials_group1 + 1 / trials_group2))

# Z-statistic formula for two proportions test: Z = (p1 - p2) / SE
z_stat = (p1 - p2) / SE

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed p-value

# Print the results
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```
