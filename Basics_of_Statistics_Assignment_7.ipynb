## Q 1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.

Ans- ### Types of Data

Data can be categorized into **qualitative** (categorical) and **quantitative** (numerical) types. These categories help in identifying how the data can be measured and analyzed.

1. **Qualitative Data (Categorical Data)**:
   - Qualitative data consists of categories or labels that describe qualities or characteristics.
   - This type of data cannot be measured numerically and is often represented by words, labels, or other non-numeric symbols.
   - Examples:
     - **Color of a car** (Red, Blue, Green)
     - **Gender** (Male, Female, Non-binary)
     - **Favorite sport** (Football, Basketball, Tennis)
   
   Qualitative data can be further classified into **nominal** and **ordinal** scales:
   
   - **Nominal Data**:
     - The simplest form of data where the categories do not have any meaningful order or ranking.
     - The categories are mutually exclusive, meaning each item can belong to only one category.
     - Examples:
       - **Marital Status** (Single, Married, Divorced, Widowed)
       - **Blood Type** (A, B, AB, O)
       - **Eye Color** (Brown, Blue, Green, Hazel)

   - **Ordinal Data**:
     - Data that can be categorized into distinct categories with a meaningful order or ranking.
     - However, the difference between the categories is not uniform or measurable.
     - Examples:
       - **Education Level** (High School, Bachelor's, Master's, PhD)
       - **Rating Scales** (1 star, 2 stars, 3 stars, 4 stars, 5 stars for a hotel rating)
       - **Class Rankings** (First, Second, Third)

2. **Quantitative Data (Numerical Data)**:
   - Quantitative data is numerical and can be measured and expressed in terms of numbers.
   - It can be used to perform arithmetic operations like addition, subtraction, multiplication, and division.
   - Examples:
     - **Height** (180 cm, 5’9”)
     - **Weight** (75 kg)
     - **Age** (25 years)
     - **Income** (USD 45,000 per year)

   Quantitative data can be further classified into **interval** and **ratio** scales:

   - **Interval Data**:
     - Interval data has ordered categories with consistent, equal intervals between them, but lacks a true zero point.
     - This means that differences between data points can be compared, but ratios or proportions (like twice as much) do not make sense.
     - Examples:
       - **Temperature in Celsius or Fahrenheit** (e.g., 10°C to 20°C, the difference between the values is meaningful, but 20°C is not "twice as hot" as 10°C).
       - **Calendar dates** (e.g., the year 2020 and 2025 have a 5-year difference, but we cannot say 2025 is "twice" the year 2020).

   - **Ratio Data**:
     - Ratio data has all the properties of interval data, but with the addition of a true zero point. This means that both differences and ratios between values are meaningful.
     - Examples:
       - **Height** (A person who is 180 cm tall is twice as tall as someone who is 90 cm tall)
       - **Weight** (A person weighing 100 kg weighs twice as much as one weighing 50 kg)
       - **Income** (A person earning USD 80,000 has twice the income of someone earning USD 40,000)

In summary:
- **Qualitative data** deals with labels or categories, and can be divided into nominal (no order) and ordinal (order matters but not the exact difference between categories).
- **Quantitative data** involves numerical measurements, and can be divided into interval (equal intervals but no true zero) and ratio (equal intervals and a true zero, making ratios meaningful) scales.


## Q 2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

Ans- ### Measures of Central Tendency

Measures of central tendency are statistics used to summarize a set of data by identifying the center point. The three main measures are **mean**, **median**, and **mode**. Each is useful in different situations depending on the nature of the data.

1. **Mean** (Average):
   - The sum of all data points divided by the number of points.
   - **Formula**: \( \text{Mean} = \frac{\sum X}{n} \)
   - **Example**: For test scores: 85, 90, 95, the mean = \( \frac{85 + 90 + 95}{3} = 90 \).
   - **When to Use**:
     - Best used for **interval or ratio data** that are **normally distributed** (no extreme outliers).
     - Sensitive to outliers (e.g., a very high or very low value can skew the mean).

2. **Median** (Middle Value):
   - The middle value when the data points are arranged in ascending or descending order.
   - If there's an odd number of values, the median is the middle one; if even, it’s the average of the two middle values.
   - **Example**: For scores: 85, 90, 95, the median = 90.
     - For scores: 85, 90, 95, 100, the median = \( \frac{90 + 95}{2} = 92.5 \).
   - **When to Use**:
     - Best for **ordinal, interval, or ratio data** that might contain outliers or skewed distributions (e.g., income levels, housing prices).

3. **Mode** (Most Frequent Value):
   - The value that occurs most frequently in the data set.
   - **Example**: For scores: 85, 90, 90, 95, the mode = 90 (since it appears twice).
   - **When to Use**:
     - Useful for **nominal, ordinal, interval, or ratio data**.
     - Can be used for both numerical and categorical data.
     - Best when you want to identify the most common or frequent value, especially in data with no clear pattern (e.g., most popular product, most frequent survey response).

### Summary of Usage:

- **Mean**: Use for normally distributed data without outliers.
- **Median**: Use when data is skewed or has outliers.
- **Mode**: Use when identifying the most frequent value, especially with categorical or skewed data.


## Q 3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

Ans- ### Concept of Dispersion

**Dispersion** refers to the extent to which data points in a dataset are spread out or clustered around a central value (such as the mean). It gives insight into how much variation or "spread" exists within the data. A small dispersion means the data points are close to the central value, while a large dispersion indicates that the data points are spread out over a wider range.

### Measures of Dispersion

1. **Variance**:
   - Variance measures the average squared deviation of each data point from the mean. It tells us how much each data point deviates from the mean, on average.
   - **Formula**:
     - For a sample: \( \text{Variance} (s^2) = \frac{\sum (X_i - \bar{X})^2}{n - 1} \)
     - For a population: \( \text{Variance} (\sigma^2) = \frac{\sum (X_i - \mu)^2}{N} \)
     - Where \( X_i \) is each data point, \( \bar{X} \) is the sample mean, and \( \mu \) is the population mean.
   - **Interpretation**: A higher variance indicates a greater spread of data points around the mean, while a lower variance means the data points are more concentrated near the mean.
   
   - **Example**: For scores: 4, 6, 8.
     - Mean = \( (4 + 6 + 8) / 3 = 6 \).
     - Variance = \( \frac{(4-6)^2 + (6-6)^2 + (8-6)^2}{3} = \frac{4 + 0 + 4}{3} = 2.67 \).

2. **Standard Deviation**:
   - The standard deviation is the square root of the variance. It provides a measure of dispersion in the same unit as the original data, making it more interpretable than variance.
   - **Formula**:
     - For a sample: \( \text{Standard Deviation} (s) = \sqrt{\text{Variance}} \)
     - For a population: \( \text{Standard Deviation} (\sigma) = \sqrt{\text{Population Variance}} \)
   - **Interpretation**: Like variance, a larger standard deviation indicates greater spread, but because it's in the same unit as the data, it’s easier to interpret.
   
   - **Example**: For the same data set (4, 6, 8), the standard deviation is the square root of 2.67, which is approximately **1.63**.

### Key Differences Between Variance and Standard Deviation:
- **Units**: Variance is expressed in squared units of the data, while standard deviation is in the same units as the data.
- **Interpretability**: Standard deviation is generally easier to interpret because it uses the same unit of measurement as the data points themselves.

### When to Use:
- **Variance** is useful in statistical calculations and models, especially when comparing multiple datasets.
- **Standard Deviation** is more commonly used in descriptive statistics, as it provides an easily interpretable measure of the spread of data in the same units as the data itself.


## Q 4. What is a box plot, and what can it tell you about the distribution of data?
 Ans- ### What is a Box Plot?

A **box plot** (also known as a **box-and-whisker plot**) is a graphical representation that summarizes the distribution of a dataset by showing its **minimum**, **first quartile (Q1)**, **median**, **third quartile (Q3)**, and **maximum**. It also highlights any **outliers** in the data. Box plots are particularly useful for visualizing the spread and skewness of data.

### Key Components of a Box Plot:
1. **Box**:
   - The rectangular box represents the **interquartile range (IQR)**, which is the range between the first quartile (Q1) and the third quartile (Q3). This box contains the middle 50% of the data.
   - The **line inside the box** represents the **median** (Q2), which is the middle value of the dataset.

2. **Whiskers**:
   - The "whiskers" extend from the box to the **minimum** and **maximum** values within 1.5 times the IQR from the quartiles. These whiskers give an indication of the spread of the data.
   - **Whisker Range**:
     - **Lower whisker**: Extends from Q1 to the smallest value within 1.5 times the IQR below Q1.
     - **Upper whisker**: Extends from Q3 to the largest value within 1.5 times the IQR above Q3.

3. **Outliers**:
   - Data points that fall outside the whiskers (i.e., beyond 1.5 times the IQR) are considered **outliers** and are often marked individually (e.g., with dots or asterisks).

### What Can a Box Plot Tell You?

1. **Center (Median)**:
   - The **median** line inside the box shows the central tendency of the data. If the median is closer to the bottom or top of the box, it suggests skewness in the data.

2. **Spread (Range and IQR)**:
   - The **size of the box** indicates the spread of the middle 50% of the data (IQR). A wider box suggests greater variability, while a narrow box indicates more consistent data.
   - The **whiskers** show the range of the data, with the length indicating how spread out the data is.

3. **Skewness**:
   - If the median is not in the center of the box, the data may be skewed:
     - **Right-skewed**: If the median is closer to Q1 and the right whisker is longer.
     - **Left-skewed**: If the median is closer to Q3 and the left whisker is longer.

4. **Outliers**:
   - **Outliers** are points outside the whiskers, indicating values that are significantly higher or lower than the rest of the data. Identifying outliers is important for detecting anomalies.

### Example of a Box Plot Interpretation:

Consider the following data:  
**Data**: 2, 3, 4, 5, 6, 7, 8, 9, 9, 10

A box plot for this data would show:
- **Q1 (25th percentile)**: 4
- **Q2 (Median, 50th percentile)**: 7
- **Q3 (75th percentile)**: 9
- **Minimum**: 2
- **Maximum**: 10
- **Outliers**: None (since all data points fall within the whiskers)

The box plot would show a symmetric distribution, with the median roughly in the middle of the box and equal-length whiskers. The data does not have any extreme outliers.

### Summary:
A **box plot** is a powerful tool for visually summarizing data distribution. It provides insight into the data's **central tendency**, **spread**, **skewness**, and **outliers**, making it easier to understand the underlying patterns in the dataset.


## Q 5. Discuss the role of random sampling in making inferences about populations.
 Ans- ### Role of Random Sampling in Making Inferences about Populations

**Random sampling** is a fundamental technique used in statistics to draw conclusions about a population based on data collected from a sample. By selecting a random subset of the population, random sampling ensures that every individual or item in the population has an equal chance of being included in the sample, which helps to avoid bias and increases the representativeness of the sample.

### Why Random Sampling is Important for Inferences:

1. **Representativeness**:
   - Random sampling helps ensure that the sample is **representative** of the entire population. When every individual has an equal chance of being selected, the sample is more likely to reflect the characteristics of the population accurately.
   - This reduces the risk of sampling bias, where certain groups or characteristics are overrepresented or underrepresented, leading to inaccurate conclusions.

2. **Generalization**:
   - By using random sampling, researchers can **generalize** the findings from the sample to the broader population. Since the sample is likely to reflect the population's diversity, conclusions drawn from the sample can be applied with greater confidence to the entire population.
   - For example, if a random sample of 1,000 voters in a country shows a preference for a political candidate, the result can be generalized to the whole electorate.

3. **Reduction of Bias**:
   - Random sampling minimizes bias that can arise from non-random selection methods (e.g., convenience sampling, where individuals are chosen based on ease of access). This is crucial in ensuring that the sample does not systematically favor certain groups or outcomes, which would distort the conclusions.

4. **Unbiased Estimators**:
   - Random sampling provides **unbiased estimators** for key population parameters (e.g., mean, proportion). This means that the sample statistics are expected to be close to the true population values on average, making the sample a reliable source for inference.
   - For example, the mean of a random sample is an unbiased estimator of the population mean.

5. **Foundation for Statistical Inference**:
   - Random sampling is the basis for many **statistical inference techniques** like hypothesis testing, confidence intervals, and regression analysis. These methods rely on the assumption that the sample is representative of the population, allowing for accurate estimation of population parameters and testing of hypotheses.
   
6. **Control of Variability**:
   - Since random sampling involves selecting data points without a predefined pattern, it helps control **sampling variability**. Over repeated samples, random sampling leads to a distribution of sample statistics that can be used to estimate population parameters with a certain level of confidence.

### Example of Random Sampling and Inference:

Imagine a university wants to know the average GPA of all its students. Since it's impractical to survey every student, they randomly select 200 students from a total population of 10,000.

- **Step 1: Random Sampling**: By using a random selection method (e.g., computer-generated random numbers), they ensure that each student has an equal chance of being chosen.
  
- **Step 2: Analyze the Sample**: The sample of 200 students has an average GPA of 3.5.

- **Step 3: Make Inferences**: Because of random sampling, they can **infer** that the average GPA for the entire student population is approximately 3.5, with a certain margin of error (based on the sample size and variability).

### Key Considerations in Random Sampling:
- **Sample Size**: Larger sample sizes tend to produce more reliable inferences because they reduce the sampling error (the difference between the sample statistic and the true population parameter).
- **Sampling Method**: Random sampling should be done properly (e.g., simple random sampling, stratified random sampling) to avoid potential pitfalls like unintentional bias or misrepresentation.

### Conclusion:
Random sampling plays a crucial role in making valid inferences about a population. By selecting a sample that represents the population, random sampling allows for generalizations, reduces bias, and supports statistical methods that are fundamental for analyzing data and drawing conclusions about the larger group.


## Q 6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

Ans- ### Concept of Skewness

**Skewness** refers to the asymmetry or lack of symmetry in the distribution of data. It describes the direction and degree to which a dataset deviates from a normal (bell-shaped) distribution. A distribution can be **positively skewed** (right-skewed) or **negatively skewed** (left-skewed), depending on which tail is longer or more pronounced.

### Types of Skewness:

1. **Positive Skew (Right Skew)**:
   - The **right tail** (larger values) is longer than the left.
   - The **mean** is greater than the **median** because the few large values pull the mean to the right.
   - **Example**: Income distribution, where a few people earn significantly more than the rest.

2. **Negative Skew (Left Skew)**:
   - The **left tail** (smaller values) is longer than the right.
   - The **mean** is less than the **median** because the few small values pull the mean to the left.
   - **Example**: Age of retirement, where most people retire around the same age, but a few retire much earlier.

3. **No Skew (Symmetric Distribution)**:
   - The distribution is **balanced** with both tails being equally distributed.
   - The **mean** and **median** are approximately the same.
   - **Example**: A normal distribution (e.g., test scores in a well-designed exam).

### How Skewness Affects the Interpretation of Data:

- **Measures of Central Tendency**: Skewness affects the relationship between the mean, median, and mode. In a positively skewed distribution, the mean is greater than the median. In a negatively skewed distribution, the mean is smaller than the median.
  
- **Data Analysis**: Skewed data may require transformation (e.g., log transformation) to make it more normally distributed, especially for statistical techniques that assume normality.

- **Decision Making**: When data is skewed, relying solely on the mean might mislead interpretations. The median might be a better measure of central tendency for skewed data, as it is less affected by outliers.


## Q 7. What is the interquartile range (IQR), and how is it used to detect outliers?

Ans- ### Interquartile Range (IQR)

The **Interquartile Range (IQR)** is the range between the **first quartile (Q1)** and the **third quartile (Q3)**, representing the middle 50% of a dataset. It is calculated as:

\[
\text{IQR} = Q3 - Q1
\]

Where:
- **Q1 (25th percentile)**: The value below which 25% of the data fall.
- **Q3 (75th percentile)**: The value below which 75% of the data fall.

### IQR and Outliers

Outliers are typically identified using the IQR method, based on values that fall outside a certain range. Outliers are data points that lie more than **1.5 times the IQR** above Q3 or below Q1.

- **Lower Bound**: \( Q1 - 1.5 \times \text{IQR} \)
- **Upper Bound**: \( Q3 + 1.5 \times \text{IQR} \)

Any data points below the lower bound or above the upper bound are considered **outliers**.

### Example:
For a dataset:
- Q1 = 10, Q3 = 30, IQR = 30 - 10 = 20
- Lower Bound = 10 - (1.5 × 20) = -20
- Upper Bound = 30 + (1.5 × 20) = 60

Data points below -20 or above 60 would be outliers.

### Summary:
The IQR is a measure of data spread, and it helps in detecting outliers by identifying values that fall significantly outside the typical range of the data.


## Q 8. Discuss the conditions under which the binomial distribution is used.

Ans- ### Conditions for Using the Binomial Distribution

The **binomial distribution** is used to model situations where there are a fixed number of trials, each with two possible outcomes (success or failure). The distribution is based on the following key conditions:

1. **Fixed Number of Trials (n)**:
   - The experiment must be repeated a specific number of times. Each repetition is called a trial.
   - Example: Tossing a coin 10 times.

2. **Two Possible Outcomes**:
   - Each trial has only two possible outcomes: "success" (e.g., heads, win) or "failure" (e.g., tails, loss).
   - Example: Rolling a die and checking whether the result is a 6 (success) or not (failure).

3. **Constant Probability of Success (p)**:
   - The probability of success (denoted as **p**) remains the same for each trial.
   - Example: In a series of coin flips, the probability of getting heads (success) is always 0.5.

4. **Independence of Trials**:
   - The trials must be independent, meaning the outcome of one trial does not affect the outcome of another.
   - Example: One coin flip does not influence the outcome of the next flip.

5. **Discrete Random Variable**:
   - The random variable of interest is the **number of successes** in the fixed number of trials.
   - Example: Counting how many times heads appears in 10 flips of a coin.

### Binomial Distribution Formula:
The binomial distribution is often represented by the formula:

\[
P(X = k) = \binom{n}{k} p^k (1 - p)^{n-k}
\]

Where:
- \( P(X = k) \) is the probability of exactly **k successes** in **n trials**.
- \( \binom{n}{k} \) is the binomial coefficient (combinations of choosing k successes from n trials).
- \( p \) is the probability of success on a single trial.
- \( 1 - p \) is the probability of failure.

### Example:
- If you flip a fair coin 5 times (n = 5) and want to know the probability of getting exactly 3 heads (k = 3), the binomial distribution would be used with **p = 0.5** (probability of heads).

### Summary:
The binomial distribution is used when there are a fixed number of independent trials, each with two outcomes, a constant probability of success, and the goal is to find the probability of a certain number of successes.


## Q 9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

Ans- ### Properties of the Normal Distribution

The **normal distribution** is a continuous probability distribution that is symmetric and bell-shaped. Key properties include:

1. **Symmetry**: The distribution is perfectly symmetrical around its mean. The left and right sides of the curve are mirror images.
2. **Mean, Median, Mode**: In a normal distribution, the **mean**, **median**, and **mode** are all equal and located at the center of the distribution.
3. **Bell-Shaped Curve**: The curve gradually tapers off as it moves away from the mean, with most of the data clustered around the center.
4. **Asymptotic**: The tails of the distribution approach but never touch the horizontal axis. This means there is always a small probability of extreme values.
5. **Defined by Mean and Standard Deviation**: The shape of the normal distribution is determined by two parameters: the **mean (μ)** and the **standard deviation (σ)**. The mean defines the center, while the standard deviation controls the spread.

### The Empirical Rule (68-95-99.7 Rule)

The **Empirical Rule** describes the percentage of data that falls within certain standard deviations of the mean in a normal distribution:

- **68%** of the data falls within **1 standard deviation (σ)** of the mean (μ).
- **95%** of the data falls within **2 standard deviations (2σ)** of the mean.
- **99.7%** of the data falls within **3 standard deviations (3σ)** of the mean.

### Example:
For a normal distribution with a mean of 50 and a standard deviation of 10:
- **68%** of the data is between 40 and 60 (μ ± 1σ).
- **95%** of the data is between 30 and 70 (μ ± 2σ).
- **99.7%** of the data is between 20 and 80 (μ ± 3σ).

### Summary:
The **normal distribution** is symmetric and bell-shaped, and its behavior can be predicted using the **Empirical Rule**, which shows that most of the data lies within a few standard deviations of the mean.


## Q 10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.

Ans- ### Real-Life Example of a Poisson Process:

A **Poisson process** models events that occur **randomly and independently** over a fixed period of time or space, with a known average rate (λ).

#### Example: Traffic Flow at a Toll Booth
Suppose cars arrive at a toll booth at an average rate of **5 cars per hour** (λ = 5). The question could be: What is the probability that exactly **3 cars** will arrive in the next hour?

### Poisson Distribution Formula:

The probability of observing exactly **k** events in a given time period is given by the Poisson distribution formula:

\[
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
\]

Where:
- \( \lambda \) is the average rate of events (5 cars per hour in this case).
- \( k \) is the number of events we are interested in (3 cars).
- \( e \) is Euler's number (approximately 2.71828).

### Calculation:

For \( k = 3 \) and \( \lambda = 5 \):

\[
P(X = 3) = \frac{5^3 e^{-5}}{3!} = \frac{125 \times e^{-5}}{6} = \frac{125 \times 0.006737}{6} \approx 0.1404
\]

### Result:

The probability that exactly **3 cars** will arrive at the toll booth in the next hour is approximately **0.1404** or **14.04%**.

This is an example of a **Poisson process** where the number of events (car arrivals) follows a Poisson distribution, and we used the formula to calculate the probability of a specific number of events occurring in a fixed time period.


## Q 11. Explain what a random variable is and differentiate between discrete and continuous random variables.

Ans- ### What is a Random Variable?

A **random variable** is a numerical outcome of a random phenomenon or experiment. It is a variable whose value is determined by the outcome of a random event, and it can take on different values, each with a certain probability.

### Types of Random Variables:

1. **Discrete Random Variable**:
   - A discrete random variable can only take **specific, distinct values**, often integers or counts.
   - **Examples**:
     - The number of heads in 5 coin flips.
     - The number of customers arriving at a store in an hour.
   - The possible values are countable and typically finite.

2. **Continuous Random Variable**:
   - A continuous random variable can take **any value** within a given range or interval, and it can have an infinite number of possible outcomes.
   - **Examples**:
     - The height of a person.
     - The time it takes to complete a task.
   - The values are not countable but are measured on a continuous scale.

### Summary:
- **Discrete Random Variable**: Takes specific, countable values (e.g., number of children, number of cars).
- **Continuous Random Variable**: Takes any value within a range (e.g., weight, temperature).


## Q 12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.

Ans- ### Example Dataset

Consider the following dataset of **X** (Hours Studied) and **Y** (Test Scores):

| Hours Studied (X) | Test Score (Y) |
|-------------------|----------------|
| 1                 | 50             |
| 2                 | 55             |
| 3                 | 60             |
| 4                 | 65             |
| 5                 | 70             |

### Step 1: Calculate Covariance

Covariance measures the degree to which two variables change together. The formula for covariance is:

\[
\text{Cov}(X, Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n}
\]

#### Step 1a: Calculate the Means of X and Y

- \(\bar{X} = \frac{1 + 2 + 3 + 4 + 5}{5} = 3\)
- \(\bar{Y} = \frac{50 + 55 + 60 + 65 + 70}{5} = 60\)

#### Step 1b: Calculate Covariance

| X  | Y  | \(X_i - \bar{X}\) | \(Y_i - \bar{Y}\) | \((X_i - \bar{X})(Y_i - \bar{Y})\) |
|----|----|-------------------|-------------------|------------------------------------|
| 1  | 50 | -2                | -10               | 20                                 |
| 2  | 55 | -1                | -5                | 5                                  |
| 3  | 60 | 0                 | 0                 | 0                                  |
| 4  | 65 | 1                 | 5                 | 5                                  |
| 5  | 70 | 2                 | 10                | 20                                 |

\[
\sum (X_i - \bar{X})(Y_i - \bar{Y}) = 20 + 5 + 0 + 5 + 20 = 50
\]

\[
\text{Cov}(X, Y) = \frac{50}{5} = 10
\]

### Step 2: Calculate Correlation

The formula for Pearson correlation coefficient \(r\) is:

\[
r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
\]

Where \(\sigma_X\) and \(\sigma_Y\) are the standard deviations of \(X\) and \(Y\).

#### Step 2a: Calculate Standard Deviations

- \(\sigma_X = \sqrt{\frac{\sum (X_i - \bar{X})^2}{n}} = \sqrt{\frac{10}{5}} = \sqrt{2} \approx 1.414\)
- \(\sigma_Y = \sqrt{\frac{\sum (Y_i - \bar{Y})^2}{n}} = \sqrt{\frac{250}{5}} = \sqrt{50} \approx 7.071\)

#### Step 2b: Calculate Correlation

\[
r = \frac{10}{1.414 \times 7.071} = \frac{10}{10} = 1
\]

### Interpretation of Results

1. **Covariance** = 10: This indicates a **positive relationship** between hours studied and test scores. As the number of hours studied increases, the test scores also increase.

2. **Correlation** = 1: A correlation of **1** indicates a **perfect positive linear relationship** between the two variables. This means that test scores increase in a perfectly predictable way as the number of hours studied increases.

### Conclusion:

The results show a strong, positive linear relationship between the hours studied and test scores. As the hours studied increase, the test scores rise in a consistent and predictable manner.
